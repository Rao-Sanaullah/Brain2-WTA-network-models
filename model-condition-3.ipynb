```python
# List of Condition: Fully connected network

import time

start = time.time()
"the code you want to test stays here"

%matplotlib inline
from brian2 import *

# Visualise the connectivity of a connection between two Populations
def visualise_connectivity(S):
    Ns = len(S.source)
    Nt = len(S.target)
    figure(figsize=(10, 4))
    subplot(121)
    plot(zeros(Ns), arange(Ns), 'ok', ms=10)
    plot(ones(Nt), arange(Nt), 'ok', ms=10)
    for i, j in zip(S.i, S.j):
        plot([0, 1], [i, j], '-k')
    xticks([0, 1], ['Source', 'Target'])
    ylabel('Neuron index')
    xlim(-0.1, 1.1)
    ylim(-1, max(Ns, Nt))
    subplot(122)
    plot(S.i, S.j, 'ok')
    xlim(-1, Ns)
    ylim(-1, Nt)
    xlabel('Source neuron index')
    ylabel('Target neuron index')



duration = 1*second  # Duration of the simulation
v_rest = -65*mV   # Resting Potential
c_m = 0.2*nF      # Membrance Capacitance
vt = -55*mV       # Threshold Potential
vr = -90*mV       # Reset Potential
tr = 5*ms         # Refractory Time
tau_m = 20*ms      # Membrane time constant
g_l = c_m/tau_m   # Leak conductance
noise_rate = 50 * Hz
n_neurons = 100
eqs = '''
dv/dt = (-(v-v_rest)*g_l + i_e)/c_m : volt (unless refractory)
di_e/dt = -i_e/tau_e : ampere
'''
tau_e = 2*ms

group = NeuronGroup(n_neurons, eqs, threshold='v > vt', reset='v = vr',refractory=tr, method='rk2')

spikemon = SpikeMonitor(group)

input_group = PoissonGroup(n_neurons, rates=noise_rate)



S = Synapses(group, group, model='w : ampere', on_pre='i_e += w')
S.connect(condition='i != j or i == j')


visualise_connectivity(S)

end = time.time()
print(end - start)
```
